#### <b>What is Customer Churn?</b>
<span style="font-size:16px;">  Customer churn occurs when customers or subscribers stop doing business with a company or discontinue using its services. In the telecom industry, where customers have numerous service providers to choose from, the competition is fierce, and churn rates range from 15% to 25% annually. Customers frequently switch providers, making retention a critical challenge. </span>

<span style="font-size:16px;"> Customers in the Streaming industry can choose from a variety of service providers and actively switch from one to the next. The telecommunications business has an annual churn rate of 15-25 percent in this highly competitive market.</span>

<span style="font-size:16px;"> Given the vast customer base most telecom companies manage, personalizing retention efforts for every individual isnâ€™t feasible. The costs of such efforts would quickly surpass the potential revenue gains. However, if a company could predict which customers are at high risk of leaving, it could focus its retention strategies on those specific customers. By targeting only the "high risk" individuals, the company can maximize its retention efforts more efficiently.The ultimate objective is to grow customer loyalty and reduce churn, which is vital in a highly competitive landscape. Retaining existing customers is far more cost-effective than acquiring new ones, making customer churn a key performance indicator for sustained success in the telecom sector. At its core, success in this market lies in understanding and prioritizing the customer.
</span>

<span style="font-size:16px;"> Customer churn is a critical metric because it is much less expensive to retain existing customers than it is to acquire new customers.</span>

<a id="churn"></a>
<a id = "3" ></a>

<span style="font-size:16px;"><b>To reduce customer churn, telecom companies need to predict which customers are at high risk of churn.</b></span> 

<span style="font-size:16px;"> To detect early signs of potential churn, one must first develop a holistic view of the customers and their interactions across numerous channels, including store/branch visits, product purchase histories, customer service calls, Web-based transactions, and social media interactions, to mention a few. </span> 

<span style="font-size:16px;">As a result, by addressing churn, these businesses may not only preserve their market position, but also grow and thrive. More customers they have in their network, the lower the cost of initiation and the larger the profit. As a result, the company's key focus for success is reducing client attrition and implementing effective retention strategy. </span>

#### **Voting Classifier**
<a id = "vc" ></a>
Let's now predict the final model based on the highest majority of voting and check it's score.

```python
# Instantiate the classifiers
clf1 = DecisionTreeClassifier()
clf2 = LogisticRegression()
clf3 = RandomForestClassifier()
clf4 = KNeighborsClassifier()
# The SVC classifier does not support the predict_proba method by default, which is required for soft voting in the VotingClassifier
clf5 = SVC(probability=True) # Enable probability estimates

# Create the VotingClassifier
eclf1 = VotingClassifier(estimators=[('dt', clf1), ('lr', clf2), ('rf', clf3), ('knn', clf4), ('svc', clf5)], voting='soft')

# Fit the model
eclf1.fit(X_train, y_train)

# Make predictions
predictions = eclf1.predict(X_test)

# Print the accuracy score
print("Final Accuracy Score ")
print(accuracy_score(y_test, predictions))
```

From the confusion matrix we can see that:
